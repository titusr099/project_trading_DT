```python
import yfinance as yf
import pandas as pd
```

```python
# Descargar datos históricos para BTC
#Y,m,d
btc = yf.download('BTC-USD', start='2020-12-24', end='2025-04-12')
```

```python
btc.to_csv('../../data/raw/BTC-USD.csv')
```

```python
btc.info()
```

```python
btc.columns = ['Close', 'High', 'Low', 'Open', 'Volume']
```

```python
btc
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
btc.describe()
```

```python
btc.isnull().sum()
```

```python
# Histograma de precio de cierre
plt.figure(figsize=(8,6))
sns.histplot(btc['Close'], bins=50, kde=True)
plt.title('Distribución del Precio de Cierre (BTC)')
plt.xlabel('Precio de Cierre (USD)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Serie temporal del precio de cierre
plt.figure(figsize=(14,7))
btc['Close'].plot()
plt.title('Precio de cierre diario de Bitcoin (BTC-USD)')
plt.ylabel('Precio en USD')
plt.xlabel('Fecha')
plt.grid(True)
plt.show()
```

```python
# Matriz de correlación
plt.figure(figsize=(10,6))
sns.heatmap(btc[['Open', 'High', 'Low', 'Close', 'Volume']].corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de correlación')
plt.show()
```

```python
btc['daily_return'] = btc['Close'].pct_change()
```

```python
# Calcula y visualiza retornos diarios
plt.figure(figsize=(12,6))
sns.histplot(btc['daily_return'].dropna(), bins=50, kde=True, color='purple')
plt.title('Distribución de los Retornos Diarios')
plt.xlabel('Retorno diario')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Serie temporal retornos diarios
plt.figure(figsize=(14,7))
btc['daily_return'].plot()
plt.title('Retornos diarios BTC-USD')
plt.ylabel('Retorno')
plt.xlabel('Fecha')
plt.grid(True)
plt.show()
```

```python
# Promedio de retorno diario por día de la semana
btc['day_of_week'] = btc.index.dayofweek

weekly_return = btc.groupby('day_of_week')['daily_return'].mean()

plt.figure(figsize=(10,6))
weekly_return.plot(kind='bar', color='orange')
plt.title('Retorno promedio por día de la semana')
plt.ylabel('Retorno promedio')
plt.xlabel('Día de la semana (0=Lunes)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=btc['daily_return'])
plt.title('Boxplot de Retornos Diarios (detección rápida de outliers)')
plt.xlabel('Retorno diario')
plt.show()
```

```python
Q1 = btc['daily_return'].quantile(0.25)
Q3 = btc['daily_return'].quantile(0.75)
IQR = Q3 - Q1

# Definir un rango más estricto, por ejemplo 2.5*IQR en vez de 1.5*IQR
limite_inferior = Q1 - 3.5 * IQR
limite_superior = Q3 + 3.5 * IQR

btc_filtrado = btc[(btc['daily_return'] >= limite_inferior) & (btc['daily_return'] <= limite_superior)]
```

```python
btc_filtrado['daily_return'].describe()
plt.figure(figsize=(10,5))
sns.boxplot(x=btc_filtrado['daily_return'])
plt.title('Boxplot de Retornos Diarios (sin outliers)')
plt.xlabel('Retorno diario')
plt.show()
```

```python
eliminados = len(btc) - len(btc_filtrado)
print(f"Se eliminaron {eliminados} datos extremos.")
```

```python
porcentaje = (eliminados / len(btc)) * 100
print(f"Se eliminó el {porcentaje:.2f}% de los datos.")
```
